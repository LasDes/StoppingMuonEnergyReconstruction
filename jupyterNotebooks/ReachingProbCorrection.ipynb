```python
r_icecube = 800
depth_icecube = 1950

def calc_tau(l):
    return np.arctan(depth_icecube / l)

def calc_Theta(l):
    d = np.sqrt(l**2 + depth_icecube**2)
    return 2 * np.arctan(r_icecube / d)

def stoppoint(r, l, theta, phi, tau):
    h = np.cos(theta/2) * np.sin(tau) * r
    t = np.cos(theta/2) * np.cos(tau) * r
    x = l - t - np.cos(phi) * np.sin(theta/2) * np.sin(theta/2) * r
    y = depth_icecube - h + np.cos(phi) * np.sin(theta/2) * np.cos(theta/2) * r
    # x = l + np.cos(phi) * np.sin(theta/2) * np.sin(theta/2) * r
    # y = h - np.cos(phi) * np.sin(theta/2) * np.cos(theta/2) * r
    z = np.sin(phi) * np.sin(theta/2) * r
    return (x,y,z)

def solid_angle(Theta):
    return 4 * np.pi * np.sin(Theta/4)**2

def bin_stop_prob(ranges, steps, l_min, l_max):
    np.random.seed(seed=666)
    result = 0
    bin_right = np.linspace(l_min, l_max, num=steps)
    bin_left = np.array([0.0] + list(bin_right[:-1]))
    bin_width = bin_right - bin_left
    for i, l in enumerate(bin_right):
        Theta = calc_Theta(l)
        tau = calc_tau(l)
        theta = Theta * np.random.ranf(100000)
        phi = 2 * np.pi * np.random.ranf(100000)
        data = pd.DataFrame({'zenith': np.pi/2 - tau + np.arctan(np.cos(phi) * np.tan(theta))})
        data['zeta'] = np.arctan(np.cos(phi) * np.tan(theta))
        data['zen_cos'] = np.cos(data.zenith)
        zen_bins = np.linspace(np.cos(79.2 / 180 * np.pi), 1, num=11)
        data['zen_index'] = pd.cut(data.zen_cos, zen_bins)
        locations = stoppoint(ranges, l, theta, phi, tau)
        data['stopping'] = det_hull.find_simplex(zip(*locations)) > 0
        data.dropna(inplace=True)
        data['effective_area'] = area_byR.effective_area.loc[data.zen_index].values
        data['weight'] = 1 / data['effective_area']
        result += data[data.stopping].weight.sum() / data.weight.sum() * (solid_angle(Theta) / (4 * np.pi)) * bin_width[i]
    return result
```

```python
results = []
for i in np.arange(1,11):
    results += [bin_stop_prob(ranges[ranges.I == i].R, 500, 0.1, 15000.)]
print results
```

```python
plt.plot(range(1, 11), np.array(results))
plt.xlabel('Bin')
plt.xlim(1,10)
```
